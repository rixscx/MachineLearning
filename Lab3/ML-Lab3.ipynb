# LAB 3: Develop a program to implement Principal Component Analysis (PCA) for reducing the dimensionality of the Iris dataset from 4 features to 2. 

```python
import pandas as pd 
from sklearn.decomposition import PCA 
import matplotlib.pyplot as plt
```

## Read the Iris dataset 

```python
data = pd.read_csv("Iris.csv") 
```

## Display basic information about the dataset

```python
print(data.info())
print("\nFirst 5 Rows:")
print(data.head())
```

## Select the 4 feature columns 

```python
X = data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
```

## Apply PCA to reduce to 2 components 

```python
pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X) 
```

## Print the shape and the first few rows of the reduced data 

```python
print("Reduced data shape:", X_pca.shape) 
print(X_pca[:5]) 
```

## Plot the 2D PCA result 

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1]) 
plt.xlabel("Principal Component 1") 
plt.ylabel("Principal Component 2") 
plt.title("PCA of Iris Dataset") 
plt.show() 
```
